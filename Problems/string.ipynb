```python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        s = s.lower()
        for i in s:
            if i.isalpha() or i.isdigit():
                pass
            else:
                s = s.replace(i, '')
                
        return s == s[::-1]
        
        
        
Solution().isPalindrome(" ")
```

```python
# Length of Last Word

class Solution(object):
    def lengthOfLastWord(self, s:str):
        """
        :type s: str
        :rtype: int
        """
        s = s.strip()
        slist = s.split(" ")
        
        return len(slist[-1])
    
Solution().lengthOfLastWord("Hello World")
        
```
