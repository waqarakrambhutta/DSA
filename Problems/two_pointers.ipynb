```python
from typing import List
```

```python

class Solution:
    def reverseString(self, s):

        """
        Do not return anything, modify s in-place instead.
        """
        for i in range(len(s)//2):
            s[i], s[len(s)-i-1] = s[len(s)-i-1], s[i]
        
        return s
        
Solution().reverseString(["H","a","n","n","a","h"])
```
