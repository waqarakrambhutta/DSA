```python

class Solution:
    def reverseString(self, s):

        """
        Do not return anything, modify s in-place instead.
        """
        for i in range(len(s)//2):
            s[i], s[len(s)-i-1] = s[len(s)-i-1], s[i]
        
        return s
        
Solution().reverseString(["H","a","n","n","a","h"])
```

```python
class Solution:
    def isPalindrome(self, s):
        if s == s[::-1]:
            return True
        return False
    
    def validPalindrome(self, s: str) -> bool:
        for i in range(len(s)//2):
            if s[i] == s[len(s)-i-1]:
                pass
            else:
                if s[i+1] == s[len(s)-i-1] and s[i] == s[len(s)-i-1-1]:
                    if (self.isPalindrome(s[:i] + s[i+1:])):
                        return True
                    elif (self.isPalindrome(s[:len(s)-i-1] + s[len(s)-i-1+1:])):
                        return True
                    else:
                        pass
                elif s[i+1] == s[len(s)-i-1]:
                    s = s[:i] + s[i+1:]
                elif s[i] == s[len(s)-i-1-1]:
                    s = s[:len(s)-i-1] + s[len(s)-i-1+1:]
                break
            
        return self.isPalindrome(s)
    # tcaac
    # eccer
Solution().validPalindrome("eedede")
```

```python
class Solution:
    def mergeAlternately(self, word1: str, word2: str) -> str:
        l1 = len(word1)
        l2 = len(word2)
        i = l1
        if l1 > l2:
            i = l2
        result = ""
        for i in range(i):
            result += word1[i] + word2[i]        
        if i == l2-1:
            result += word1[i+1:]
        else:
            result += word2[i+1:]
        return result

Solution().mergeAlternately("ab","pqrs")
```
