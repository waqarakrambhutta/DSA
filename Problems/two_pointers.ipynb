```python
from typing import List
```

```python
# Reverse String

class Solution:
    def reverseString(self, s: List[str]) -> None:

        """
        Do not return anything, modify s in-place instead.
        """
        for i in range(len(s)):
            s.insert(0, s[-1])

        return s
        
    
Solution().reverseString(["h","e","l","l","o"])
```
