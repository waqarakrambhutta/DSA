```python
text = "hello world"
number = 123
float_num = 45.67
vowels_tuple = ('a', 'e', 'i', 'o', 'u')

print(f'Hash for "{text}" is: {hash(text)}')
print(f'Hash for {number} is: {hash(number)}')
print(f'Hash for {float_num} is: {hash(float_num)}')
print(f'Hash for {vowels_tuple} is: {hash(vowels_tuple)}')

try:
    hash([1, 2, 3])
except TypeError as e:
    print(f"Error hashing a list: {e}")
```

```python
class MyHashSet:

    def __init__(self):
        self.hash = {}
        
    def __str__(self):
        return str(self.hash)

    def add(self, key: int) -> None:
        self.hash[key] = key
        
    def remove(self, key: int) -> None:
        if key in self.hash:
            self.hash.pop(key)
        
    def contains(self, key: int) -> bool:
        return key in self.hash
        
        


myHashSet = MyHashSet()

myHashSet.add(1);      # set = [1]
myHashSet.add(2);      # set = [1, 2]
myHashSet.contains(1); # return True
myHashSet.contains(3); # return False, (not found)
myHashSet.add(2);      # set = [1, 2]
myHashSet.contains(2); # return True
myHashSet.remove(2);   # set = [1]
myHashSet.contains(2); # return False, (already removed)


print(myHashSet)
```
