```python
# Contain Duplication

class Solution(object):
    def containsDuplicate(self, nums):
        """
        :type nums: List[int]
        :rtype: bool
        """
        
        return len(nums) != len(set(nums))
                

print(Solution().containsDuplicate([2,14,18,22]))
```

```python
class Solution(object):
    def isAnagram(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: bool
        """
        if len(s) != len(t):
            return False

        for i in set(s):
            if s.count(i) != t.count(i):
                return False
        
        return True

Solution().isAnagram('anagram', 'nagaram')
```

```python
class Solution(object):
    def groupAnagrams(self, strs):
        """
        :type strs: List[str]
        :rtype: List[List[str]]
        """
        my_dict = {}
        # my_dict['ate'] = ['eat']
        # print(my_dict['eat'])

        for i in strs:
            key = ''.join(sorted(i))
                    
            if key in my_dict:
                my_dict[key].append(i)
            else:
                my_dict[key] = [i]

        return my_dict.values()





            
Solution().groupAnagrams(["eat","tea","tan","ate","nat","bat"])
```

```python
# Top K Frequent Elements
class Solution(object):
    def topKFrequent(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: List[int]
        """
        occurence = {}
        for i in set(nums):
            occurence[i] = nums.count(i)

        print(sorted(occurence.items(),reverse=True, key=lambda item: item[1]))

        return list(dict(sorted(occurence.items(),reverse=True, key=lambda item: item[1])).keys())[:k]
    

            

Solution().topKFrequent([4,1,-2,2,-2,2,3],2)
        

```

```python
# Product of Array Except Self

from array import * 

class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:
        tmplt = [1] * len(nums)
        def multiple(list: List):
            product = 1
            for i in list:
                product *= i
            return product

        for i in range(len(nums)):
            tmplt[i] = multiple(nums[:i]) * multiple(nums[i+1:])

        return tmplt

        
            
       
        
Solution().productExceptSelf([-1,1,0,-3,3])

```

```python
# Single Number

class Solution:
    def singleNumber(self, nums: List[int]) -> int:
        gate = 0
        for i in nums:
            gate ^= i
            
        return gate
            
Solution().singleNumber([4,1,2,1,2])
        
```

```python
# Concatination of Array.

class Solution:
    def getConcatenation(self, nums: List[int]) -> List[int]:
        n = len(nums)
        ans = []
        for i in range(n*2):
            if i < n:
                ans.append(nums[i])
            else:
                ans.append(nums[i-n])
        return ans
            
                
Solution().getConcatenation([1,2,1])
```

```python
# Remove Element

class Solution:
    def removeElement(self, nums: List[int], val: int) -> int:
        for i in range(len(nums)):
            if val in nums:
                nums.pop(nums.index(val))
            else:
                pass
        
        return len(nums)
        
        
Solution().removeElement([3,3], 3)
```

```python
class Solution:
    def mergeSorted(self, list_one, number):
        sorted_list = []
        inserted = True
        
        for k in range(len(list_one)):            
            if list_one[k] > number and inserted: 
                list_one.insert(k, number)
                inserted = False
            else:
                pass
                
        if inserted == True:
            sorted_list.append(number)
                
        return sorted_list

        
    def sortArray(self, nums: List[int]) -> List[int]:
        result = []
        for i in nums:
            result = self.mergeSorted(result, i)
        return result
    
    

# Solution().sortArray([5,1,1,2,0,0]
Solution().mergeSorted([4,5,7], 40)
```
