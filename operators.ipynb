```python
# Bitwise Operators
# & Bitwise AND
# | Bitwise OR
# ~ Bitwise NOT
# ^ Bitwise XOR
# << Bitwise Left Shift
# >> Bitwise Right Shift
# Precedence: (Highest) ~ -> <<>> -> & -> ^ -> | (Lowest)

# Bitwise AND
a = 0b1010
b = 0b1100
c = a & b
# print(bin(c)) # Output: 0b1000

# Bitwise OR
a = 0b1010
b = 0b1100
c = a | b
# print(bin(c)) # Output: 0b1110

# Bitwise XOR
# a = 0b1010
# b = 0b1100
a = 2
b = 2
c = a ^ b
xor = 4
xor ^= 1
# print(bin(c)) # Output: 0b0110
print(bin(4))

# Bitwise NOT
a = 0b1010
c = ~a
# print(bin(c)) # Output: -0b1011

# Bitwise Left Shift
# We can say this multiply by 2 as well.
a = 0b1010
c = a << 2
# print(bin(c)) # Output: 0b101000

# Bitwise Right Shift
# We can say this divide by 2 as well.
a = 0b1010
c = a >> 1
# print(bin(c)) # Output: 0b10101

```
